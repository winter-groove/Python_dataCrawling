```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
# 타겟 사이트 주소 만들기
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg=1 => 1 ~ 50
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg=2 => 51 ~ 100
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg=3 => 101 ~ 150
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg=4 => 151 ~ 200
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg='
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    print(targetSite)
```

```python
# 노래 제목 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers=header)
    # print(request)
    html = request.text
    # print(html)
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    
    # 노래 제목은 'class' 속성이 'title ellipsis'인 <a> 태그 내부에 있다.
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(titles)
    # print(len(titles))
    
    i = 0
    for title in titles:
        i += 1
        # print(i, title.text.strip())
        # print(i, title.text.split('\n')[-1].strip())
        print(i, title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers=header)
    # print(request)
    html = request.text
    # print(html)
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    
    # 가수 이름은 'class' 속성이 'artist ellipsis'인 <a> 태그 내부에 있다.
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    # print(artists)
    # print(len(artists))
    
    i = 0
    for artist in artists[5:]:
        i += 1
        # print(i, artist.text.strip())
        # print(i, artist.text.split('\n')[0])
        print(i, artist.text.strip().split('\n')[0])
```

```python
# 크롤링한 데이터를 리스트에 저장한다.
genie = []
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210702&hh=20&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers=header)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    # genie 리스트에 노래 제목과 가수 이름을 튜플로 만들어 저장한다.
    for j in range(50):
        genie.append(
            (titles[j].text.strip().split('\n')[0], artists[j + 5].text.strip().split('\n')[0])
        )
# ===== for i in range(1, 5)
print(genie)
```

```python
print(dt.now())
for i in range(200):
    print('{0:3d}위 {1} - {2}'.format(i + 1, genie[i][1], genie[i][0]))
```

```python
# 크롤링한 결과를 텍스트 파일로 저장한다.
file = open('genieTOP200.txt', 'w')
file.write('{} 현재 Genie 뮤직 실시간 TOP 200\n'.format(dt.now()))
for i in range(200):
    data = '{0:3d}위 {1} - {2}'.format(i + 1, genie[i][1], genie[i][0])
    file.write(data + '\n')
file.close()
print('genieTOP200.txt 파일에 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 출력한다.
try:
    file = open('genieTOP200.txt', 'r')
    lines = file.readlines()
    for line in lines:
        print(line.strip())
except FileNotFoundError:
    print('genieTOP200.txt 파일이 디스크에 없습니다.')
```
