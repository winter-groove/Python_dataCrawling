```python
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(targetSite)
# print(request)
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
subjects = [] # 파싱한 대화 주제를 저장할 빈 리스트를 만든다.
divs = soup.findAll('div', {'class': 'tcb-col'})
# print(len(divs)) # 3

for div in divs:
    # print(len(div))
    chapter = div.findAll('a')
    # print(len(chapters))
    for a in chapter:
        # print(a)
        # print(a.text)
        subjects.append(a.text)
```

```python
for i in range(len(subjects)):
    print('{0:2d}. {1}'.format(i + 1, subjects[i]))
```
